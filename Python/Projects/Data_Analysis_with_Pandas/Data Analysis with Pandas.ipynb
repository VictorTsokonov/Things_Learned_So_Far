```python
df7_t
```

```python
## Adding a new Row, but addition process conducted in the column
df7_t["My Address"]=["My City","My Country",10,7,"My Shop","My State","My Continent"]
df7_t
```

```python
## Re-Switching the rows and columns to obtain orignial format
df7=df7_t.T
df7
### We get a new added Row
```

## Example -- Geocoding

```python
import geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
nom=Nominatim(user_agent="http")
```

```python
n=nom.geocode("3995 23rd St, San Francisco, CA 94114")
print(n)
```

```python
nom.geocode("735 Dolores St, San Francisco, CA 94119")
```

```python
n.latitude
```

```python
n.longitude
```

```python
type(n)
```

```python
import pandas
df=pandas.read_csv("supermarkets.csv")
df
```

```python
df["Geolocation"]=df["Address"]+" , "+df["City"]+" , "+df["State"]+" , "+df["Country"]
df
```

```python
df["Coordinates"]=df["Geolocation"].apply(nom.geocode)
df
```

```python
df.Coordinates
```

```python
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
df.Coordinates[0].longitude
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x!=None else None)
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x!=None else None)
df
```
